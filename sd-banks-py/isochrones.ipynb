# BUILDING ISOCHRONES 
### From bank and credit union points

This script uses [OpenRouteService](https://openrouteservice.org/) to build driving and walking time isochrones from all the banks and credit unions in California.

```python
import geopandas as gpd
import pandas as pd
from shapely import wkt
from shapely.geometry import Point
import fiona
from sqlalchemy import create_engine
# import sqlalchemy
# import geoalchemy2
# import psycopg2
from geopy.geocoders import Nominatim
import requests
import re
import json
```

```python
# Read in points file

b = gpd.read_file('../sql/data/processed/pts_fdic_ncua.csv')
b['pts'] = gpd.GeoSeries.from_wkt(b['pts'])
b = gpd.GeoDataFrame(b, geometry='pts')
# b.head()
```

```python
# Make a list of coordinates for processing with OpenRouteService

coords = []

for i in b.pts:
    xy = [i.x,i.y]
    coords.append(xy)
# print(coords)
```

### Driving - 60 min

```python
# Make a list of isochrones by passing each coordinate to the API
# (Not DRY because takes a long time and better to see if something goes wrong for each one)

isos60minDrive = []

for coord in coords:
    body = {
        'locations': [coord],
        'range': [3600] # Time in seconds
        }

    headers = {
        'Accept': 'application/json, application/geo+json, application/gpx+xml, img/png; charset=utf-8',
        'Authorization': 'your-api-key',
        'Content-Type': 'application/json; charset=utf-8'
    }

    call = requests.post('http://localhost:8080/ors/v2/isochrones/driving-car', json=body, headers=headers)

    print(call.status_code, call.reason) # If anything other than 200 prints, something's wrong. Can also copy the 200s count them in texteditor to get a sense of progress, how much time left. There's probaby a better way to do that ...
    # print(call.text)
    isos60minDrive.append(call.text)

# # Out put to file
# file=open('data/processed/isos60minDrive.txt','w')
# for i in isos60minDrive:
#     file.writelines([i])
# file.close()
```

```python
# Make a list of the features of each isochrone
feats = []
f = []

for iso in isos60minDrive:
    # print(type(iso))
    # print(iso)
    str = json.dumps(iso)
    dict = json.loads(iso)
    # print(type(dict))
    # print(dict)
    if 'features' in dict: # There's one iso that appears to not have any features ...
        feats.append(dict['features'][0])
    else:
        f.append(dict)

print(len(feats))
print(len(f))
print(f)

# Make a geodataframe from the list of features and export it to postgres database
gdf = gpd.GeoDataFrame.from_features(feats)
engine = create_engine("postgresql://postgres:@localhost:5432/ca-banks")  
gdf.to_postgis("iso_60_min_drive_pts_fdic_ncua", engine)
```
